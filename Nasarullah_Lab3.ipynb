# 01. Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 02. Creating arrays for x (Hours Studied) and y (Final Exam Score)

```python
x = np.array([1,2,3,4,5])
y = np.array([3,4,2,4,5])

print(f"x or Independent variable: {x}\ny or Dependent variable: {y}")
```

# 03. Finding Mean of x and y

```python
xMean = np.mean(x)
yMean = np.mean(y)
print(f"xMean: {xMean}\nyMean: {yMean}")
```

# 04. Finding Slope ""m""

```python
covXY = 0
varX = 0

n = len(x) # we need this as required to iterate n times

for i in range(n):
    covXY += (x[i] - xMean) * (y[i] - yMean)
    varX += (x[i] - xMean)**2

m = covXY / varX
print(covXY)
print(varX)
print(f"Slope: {m}")
```

# 05. Finding c or y-Intercept

```python
c = yMean - (m * xMean)
print(f"c:{c}")
```

# 06. Predicting the Value of y when x = 6, 7, 8

```python
y1 = m*6 + c
print(y)
```

```python
y2 = m*7 + c
print(y)
```

```python
y3 = m*8 + c
print(y)
```

```python
yPredict = m*x +c
yPredict
```

# 07. Plotting the Original and Predicted Values

```python
plt.scatter(x,y, color = "#e91e63", label = "Scatter")
plt.xlabel("Hours Studied")
plt.ylabel("Final Exam Score")
plt.scatter(x,yPredict, color = "#00b8d4", label = "line")
plt.legend()
plt.show()
plt.plot(x,y, label ="Scatter")
plt.plot(x,yPredict, label = "Line")
plt.legend()
plt.show()
```

# 08. Finding R Square

```python
rNum = 0
rDen = 0

for i in range(n):
    rNum += (yPredict[i] - yMean)**2
    rDen += (y[i] - yMean)**2

r = rNum / rDen

print(f"R Square: {r}")
print(f"Accuracy is: {round(r*100)}%")
```

# 1. Import Data

```python
# Data
size = [1500, 2000, 1200, 1800, 1600, 2400, 1300, 1700, 1900, 2200]
bedrooms = [3, 4, 2, 3, 2, 4, 3, 2, 3, 4]
age = [10, 5, 20, 15, 12, 8, 18, 10, 7, 4]
price = [200000, 300000, 150000, 250000, 220000, 350000, 180000, 210000, 280000, 320000]
```

# 2. Specify the Independent and Dependent variable

```python
## Independent variables
x1 = size
x2 = bedrooms
x3 = age

# Dependent variable
y = price
```

# 3. Plot the Price against Each Feature

```python
# Plotting
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].scatter(size, price)
axs[0].set_xlabel('Size (sq ft)')
axs[0].set_ylabel('Price (USD)')
axs[1].scatter(bedrooms, price)
axs[1].set_xlabel('Bedrooms')
axs[1].set_ylabel('Price (USD)')
axs[2].scatter(age, price)
axs[2].set_xlabel('Age (years)')
axs[2].set_ylabel('Price (USD)')

plt.show()
```

# 4. Find the Mean

```python
x1Mean = np.mean(x1)
x2Mean = np.mean(x2)
x3Mean = np.mean(x3)

yMean = np.mean(y)
```

# 5. Finding Slope

```python
def findSlope(x,y):
    covXY = 0
    varX = 0

    xMean = np.mean(x)
    yMean = np.mean(y)
    
    n = len(x) # we need this as required to iterate n times

    for i in range(n):
        covXY += (x[i] - xMean) * (y[i] - yMean)
        varX += (x[i] - xMean)**2

    m = covXY / varX
    # print(covXY)
    # print(varX)
    # print(f"Slope: {m}")
    m = int(m)
    return m
```

```python
# m1
```

```python
m1 = findSlope(x1,y)
print(f"Slope m1: {m1}")
```

```python
# m2
```

```python
m2 = findSlope(x2,y)
print(f"Slope m2: {m2}")
```

```python
#m3
```

```python
m3 = findSlope(x3,y)
print(f"Slope m3: {m3}")
```

# 6. Finding c or y-Intercept

```python
c = yMean - (m1 * x1Mean) - (m2 * x2Mean) - (m3 * x3Mean)
c
```

# 7. Predicting the Value of y

```python
yPredict = m1*x1 + m2*x2 + m3*x3 + c
yPredict
```

```python
rNum = 0
rDen = 0

for i in range(n):
    rNum += (yPredict[i] - yMean)**2
    rDen += (y[i] - yMean)**2

r = rNum / rDen

print(f"R Square: {r}")
print(f"Accuracy is: {round(r*100)}%")
print(f"Error: {round(100-r*100)}%")
```
